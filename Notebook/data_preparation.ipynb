## **Data Preparation**

### **Separate Music and Podcast from User's Streaming History**

#### **User's Podcast Streaming History**

```python
streaming_history_audio.info()
```

```python
# Subset of columns included in the podcast streaming history
podcast_columns = ['ts', 'ms_played', 'episode_name', 'episode_show_name', 'spotify_episode_uri', 'reason_start', 'reason_end', 'shuffle', 'skipped', 'offline', 'offline_timestamp', 'incognito_mode']
```

```python
# Create a new dataframe
streaming_history_podcast = streaming_history_audio[podcast_columns].copy()
```

```python
streaming_history_podcast.info()
```

```python
# Make sure the data only includes podcast history
streaming_history_podcast = streaming_history_podcast.dropna(subset=['spotify_episode_uri'])
```

```python
streaming_history_podcast.info()
```

```python
# Save the streaming_history_podcast data to csv
streaming_history_podcast.to_csv(r'Data\Ready to Use\streaming_history_podcast.csv')
```

#### **User's Music Streaming History**

```python
streaming_history_audio.info()
```

```python
# Make sure the data only includes music history
indices_to_drop = streaming_history_audio[streaming_history_audio['episode_name'].notnull() & streaming_history_audio['episode_show_name'].notnull() & streaming_history_audio['spotify_episode_uri'].notnull()].index
streaming_history_music = streaming_history_audio.drop(indices_to_drop)

# Remove columns related to podcast
streaming_history_music = streaming_history_music.drop(columns=['episode_name', 'episode_show_name', 'spotify_episode_uri'])
```

```python
streaming_history_music.info()
```

```python
streaming_history_music.head()
```

```python
# Save the streaming_history_music data to csv
streaming_history_music.to_csv(r'Data\Ready to Use\streaming_history_music.csv')
```

### **Merge Data from User's Streaming History and Spotify API**

```python
spotify_api_data.head()
```

```python
streaming_history_music.head()
```

```python
# Match the value of the `spotify_track_uri` column in the streaming_history_music with the value of the `track_id` column in the spotify_api_data
streaming_history_music['track_uri'] = streaming_history_music['spotify_track_uri'].str.replace('spotify:track:', '')

# Remove the `spotify_track_uri` because it's already replace with the `track_uri`
streaming_history_music = streaming_history_music.drop(columns='spotify_track_uri')
```

```python
streaming_history_music.info()
```

```python
streaming_history_music.head()
```

```python
spotify_data = streaming_history_music.merge(spotify_api_data, left_on='track_uri', right_on='track_id', how='left')
```

```python
spotify_data.info()
```

```python
spotify_data.head()
```

```python
# Save the merged data to csv
spotify_data.to_csv(r'Data\Ready to Use\spotify_data.csv')
```
